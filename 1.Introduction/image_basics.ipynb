### COLOR THRESHOLD

```python
# COLOR THRESHOLD - BLUE SCREEN


image = cv2.imread("C://Users/subha/Desktop/Computer Vision/download.png")
print(image.shape)
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)

lower_blue = np.array([0,0,230])
upper_blue = np.array([70,70,255])

mask = cv2.inRange(image_rgb, lower_blue, upper_blue)

# Mask the image to let the pizza show through
masked_image = np.copy(image_rgb)
masked_image[mask!=0]=[0,0,0]

#adding background
back = cv2.imread("C://Users/subha/Desktop/Computer Vision/image.jpg")
back = cv2.cvtColor(back,cv2.COLOR_BGR2RGB)

back = back[:192,:315]
back[mask==0] = [0,0,0]
final_image = back + masked_image
plt.imshow(final_image)
```
