## Haar Cascade - Car and Carplate Detection

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Car Detection

```python
car_cascade = cv2.CascadeClassifier('haar files/haarcascade_car.xml')

cap  = cv2.VideoCapture("Video/Vehicles.mp4")

while cap.isOpened():
    ret, frame = cap.read()
    cars = car_cascade.detectMultiScale(frame, 1.3, 2)
    
    if ret == True:
        
        # Draw bound boxes to padestrians
        for (x,y,w,h) in cars:
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.imshow("Cars",frame)
        
        if cv2.waitKey(1)==27:
            break
    
    else:
        break

# Release capture
cap.release()
cv2.destroyAllWindows()
```

## Car Plate Detection

```python
# load in color image for face detection
image = cv2.imread("C://Users/subha/Desktop/Projects/z_Resources/Computer Vision/images/car_plate.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  


plt.figure(figsize=(5,5))
plt.imshow(gray, cmap='gray')
plt.show()
```

```python
plate_cascade = cv2.CascadeClassifier('haar files/haarcascade_plate_number.xml')

plates = plate_cascade.detectMultiScale(gray)#, 1.1, 1)

img_with_detections = np.copy(image)

for (x,y,w,h) in plates:
    cv2.rectangle(img_with_detections,(x,y),(x+w,y+h),(0,255,0),3)  


plt.figure(figsize=(10,10))
plt.imshow(img_with_detections)
plt.show()
```
