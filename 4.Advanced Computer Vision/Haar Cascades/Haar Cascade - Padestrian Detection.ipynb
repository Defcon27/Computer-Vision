## Haar Cascade - Padestrians detection

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2                     
```

### From Video

```python
pades_cascade = cv2.CascadeClassifier('haar files/haarcascade_fullbody.xml')

cap  = cv2.VideoCapture("Video/People_Walking.mp4")

while cap.isOpened():
    ret, frame = cap.read()
    padestrians = pades_cascade.detectMultiScale(frame)#, 1.2, 3)
    
    if ret == True:
        
        # Draw bound boxes to padestrians
        for (x,y,w,h) in padestrians:
            cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),5)
            cv2.imshow("Pades",frame)
        
        if cv2.waitKey(1)==27:
            break
    
    else:
        break

# Release capture
cap.release()
cv2.destroyAllWindows()
```

## Image

```python
image = cv2.imread("C://Users/subha/Desktop/Projects/z_Resources/Computer Vision/images/faces.jpeg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

pades_cascade = cv2.CascadeClassifier('haar files/haarcascade_fullbody.xml')
pades = pades_cascade.detectMultiScale(gray, 1.2, 3)#, 4, 6)

img_with_detections = np.copy(image)

for (x,y,w,h) in pades:
    cv2.rectangle(img_with_detections,(x,y),(x+w,y+h),(255,0,0),20)        

plt.figure(figsize=(10,10))
plt.imshow(img_with_detections)
plt.show()
```
