### Dilation and Erosion

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#Reading a binary image
image = cv2.imread("C://Users/subha/Desktop/Computer Vision/phone.jpg",0)
plt.imshow(image,cmap="gray")

# Create a 5x5 kernel of ones
kernel = np.ones((5,5),np.uint8)

# Dilate the image
dilation = cv2.dilate(image, kernel, iterations = 1)

plt.imshow(dilation,cmap="gray")
```

```python
# Erode the image
erosion = cv2.erode(image, kernel, iterations = 1)

plt.imshow(erosion,cmap="gray")
```

```python
#opening, which is erosion followed by dilation. This is useful in noise reduction in which erosion first gets rid of noise
opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
plt.imshow(opening,cmap="gray")
```

```python
#Closing is the reverse combination of opening; it’s dilation followed by erosion, which is useful in closing small holes or dark areas within an object.
closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing,cmap="gray")
```
