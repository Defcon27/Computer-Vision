# Image Processing with scikit-image

```python
%config IPCompleter.greedy=True
# Import the modules from skimage
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color

# Load the rocket image
rocket = data.rocket()

# Convert the image to grayscale
gray_scaled_rocket = color.rgb2gray(rocket) 


# Show the grayscale image
plt.imshow(gray_scaled_rocket, cmap="gray")
```

```python
#flipping the image
flip = np.flipud(rocket)

plt.imshow(flip)
```

## Histogram of channels in image

```python
# Obtain the red channel
red_channel = rocket[:, :, 0]
green_channel = rocket[:, :, 1]
blue_channel = rocket[:, :, 2]

print(red_channel.ravel())

# Plot the red histogram with bins in a range of 256
plt.hist(red_channel.ravel(), bins=256, color='red')
plt.hist(green_channel.ravel(), bins=256, color='green')
plt.hist(blue_channel.ravel(), bins=256, color='blue')

# Set title and show
plt.title('Red Histogram')
plt.show()
```

## Thresholding Images

### Global Thresholding

```python
from skimage import io
# Import the otsu threshold function
from skimage.filters import threshold_otsu

rocket = io.imread('write.png')

# Make the image grayscale using rgb2gray
rocket_gray = color.rgb2gray(rocket)

# Obtain the optimal threshold value with otsu
thresh = threshold_otsu(rocket_gray)

# Apply thresholding to the image
binary = rocket_gray > thresh

# Show the image
plt.imshow(binary, cmap="gray")
```

### Local Thresholding

```python
# Import the local threshold function
from skimage.filters import threshold_local

page_image = io.imread('write.png')

#gray-scaling
page_gray = color.rgb2gray(page_image)

# Set the block size to 35
block_size = 35

# Obtain the optimal local thresholding
local_thresh = threshold_local(page_gray, block_size, offset=10)


# Obtain the binary image by applying local thresholding
binary_local = page_gray > (local_thresh)

# Show the binary image
print(page_gray.shape)
plt.imshow(binary_local, cmap="gray")
```

```python
# Import the try all function
from skimage.filters import try_all_threshold

# Import the rgb to gray convertor function 
from skimage.color import rgb2gray

# Turn the fruits image to grayscale
grayscale = rgb2gray(page_image)

# Use the try all method on the grayscale image
fig, ax = try_all_threshold(grayscale, verbose=False)

# Show the resulting plots
plt.show()
```
