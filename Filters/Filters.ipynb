```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2
```

## High Pass Filter

```python
image = cv2.imread("C://Users/subha/Desktop/Computer Vision/brain.png")
gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)

plt.imshow(gray,cmap='gray')


#Sobel high pass filter for horizontal and vertical edges
# 3x3 array for edge detection
sobel_y = np.array([[-1,-2,-1], 
                   [ 0, 0, 0], 
                   [ 1, 2, 1]])

sobel_x = np.array([[-1, 0,-1], 
                    [ 2, 0, 2], 
                    [-1, 0,-1]])


#Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
filtered_image1 = cv2.filter2D(gray, -1, sobel_y)
filtered_image2 = cv2.filter2D(gray, -1, sobel_x)

plt.imshow(filtered_image1, cmap='gray')
```

## Low Pass Filter before High Pass Filter

```python
# Create a Gaussian blurred image
gray_blur = cv2.GaussianBlur(gray, (9, 9), 0)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('original gray')
ax1.imshow(gray, cmap='gray')
ax2.set_title('blurred image')
ax2.imshow(gray_blur, cmap='gray')

filtered = cv2.filter2D(gray, -1, sobel_y)
filtered_blurred = cv2.filter2D(gray_blur, -1, sobel_y)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('original gray')
ax1.imshow(filtered, cmap='gray')

ax2.set_title('blurred image')
ax2.imshow(filtered_blurred, cmap='gray')
```

```python
#Applying Threshold
retval, binary_image = cv2.threshold(filtered_blurred, 50, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image, cmap='gray')
```
